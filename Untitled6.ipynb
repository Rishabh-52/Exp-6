```python
#9.Salary calculation with allowance
HRA=0
DA=0
basic=int(input("Enter the basic salary:"))

if basic<=10000:
  HRA=0.2*basic
  DA=0.8*basic
  gross_salary=HRA+DA+basic

elif basic<=20000:
  HRA=0.25*basic
  DA=0.9*basic
  gross_salary=HRA+DA+basic

else:
  HRA=0.3*basic
  DA=0.95*basic
  gross_salary=HRA+DA+basic


print("HRA",HRA)
print("DA",DA)
print("Gross salary=",gross_salary)
```
